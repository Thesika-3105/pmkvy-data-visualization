```python
import pandas as pd

df = pd.read_csv("pmkvy.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(df["State"], df["Trained"], label="Trained", alpha=0.7)
plt.bar(df["State"], df["Placed"], label="Placed", alpha=0.7)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of Candidates")
plt.title("PMKVY: Trained vs Placed Candidates by State")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df["Placed"], labels=df["State"], autopct="%1.1f%%", startangle=140)
plt.title("Placement Share by State (PMKVY)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["State"], df["Trained"], marker="o", label="Trained")
plt.plot(df["State"], df["Placed"], marker="o", label="Placed")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of Candidates")
plt.title("PMKVY: Training vs Placement Trend")
plt.legend()
plt.tight_layout()
plt.show()
```

***Key Insights:  

Training vs Placement  The bar chart showed that while training numbers are high in many states, the placement numbers are comparatively lower, highlighting the gap between training and employability.

Placement Distribution  The pie chart revealed that a few states contribute the major share of total placements, while many states have minimal impact.

Trend Analysis  The line chart clearly depicted states where placements are consistently trailing behind training, suggesting the need for targeted government intervention.

***Conclusion:

This analysis highlights that while PMKVY has successfully trained lakhs of candidates, the placement ratio is still not evenly distributed across states. This indicates a need for:

Better industry partnerships in low-placement states.

State-specific skill development policies.

Monitoring of post-training employment outcomes to ensure training leads to real career opportunities.
